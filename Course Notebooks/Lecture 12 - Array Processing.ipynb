```python
import numpy as np
```

```python
# import matplotlib pyplot library
# call matplotlib inline

import matplotlib.pyplot as plot
%matplotlib inline
```

```python
# create grid of points
points = np.arange(-5,5,.01)
dx,dy = np.meshgrid(points,points)
```

```python
# create function representation of
#   z(x,y) = sin(x) + sin(y)

z = np.sin(dx) + np.sin(dy)
```

```python
#TODO: Research the 'real' need for meshgrid
```

```python
# create plot

# title
# colorbar

plot.imshow(z)
plot.colorbar()
plot.title('z(x,y) = sin(x) + sin(y)')
    
```

```python
#list comprehension selecting A or B array values
A = [1,2,3,4]
B = [100,200,300,400]
condition = [True, True, False, False]

[(A_val if cond else B_val) for A_val, B_val,cond in zip(A,B,condition)]
```

```python
# acheive same objective using numpy where
np.where(condition,A,B)

# what is zip()?
```
